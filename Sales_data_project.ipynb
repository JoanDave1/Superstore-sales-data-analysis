```python
#Prediction of sales for the next 24 months (2 years)
#import libraries

from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, mean_absolute_error, mean_absolute_percentage_error, root_mean_squared_error
```

```python
df['order_date'] = pd.to_datetime(df['order_date'])

monthly_df = (
    df
    .groupby(pd.Grouper(key='order_date', freq='ME'))['sales']
    .sum()
    .reset_index()
    .sort_values('order_date')
)
```

```python
#Feature engineering, create lag features (sales of previous 24 months)
for i in range(1, 25):
    monthly_df[f'lag_{i}'] = monthly_df['sales'].shift(i)
```

```python
monthly_df['month'] = monthly_df['order_date'].dt.month
```

```python
monthly_df = monthly_df.dropna().reset_index(drop=True)
```

```python
X = monthly_df.drop(columns= ['sales', 'order_date'])
y = monthly_df['sales']

#Train-Test Split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, shuffle=False)

rf_model = RandomForestRegressor(n_estimators=200, random_state=42)
rf_model.fit(X_train, y_train)
```

```python
#Evaluate model

y_pred = rf_model.predict(X_test)
rmse = root_mean_squared_error(y_test, y_pred)
print("RMSE on Test Set:", rmse)
```

```python
#Predict next 24 months

last_date = monthly_df['order_date'].max()
future_dates = pd.date_range(start=last_date + pd.offsets.MonthEnd(1), periods=24,freq='ME')
future_df = pd.DataFrame({'order_date': future_dates})
future_df['month'] = future_df['order_date'].dt.month
```

```python
last_sales = list(monthly_df['sales'].iloc[-24:])
future_predictions = []

for i in range(24):
    row = {f'lag_{j+1}': last_sales[-(j+1)] for j in range(24)}
    row['month'] = future_df.loc[i, 'month']

    row_df = pd.DataFrame([row])
    prediction = rf_model.predict(row_df)[0]

    future_predictions.append(prediction)
    last_sales.append(prediction)
```

```python
future_df['predicted_sales'] = future_predictions
print(future_df)
```
